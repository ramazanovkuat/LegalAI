````python
from transformers import AutoTokenizer, AutoModel
````

````python
import torch
````

````python
model_name = 'DeepPavlov/rubert-base-cased-sentence'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
````

````python
# Функция для получения эмбеддингов предложений
def get_sentence_embedding(sentence, tokenizer, model):
    # Токенизация предложения
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        # Получение выходов модели
        outputs = model(**inputs)
    # Последний скрытый слой
    last_hidden_state = outputs.last_hidden_state
    # Усреднение по всем токенам для получения эмбеддинга предложения
    sentence_embedding = torch.mean(last_hidden_state, dim=1)
    return sentence_embedding

# Пример предложения
sentence = "Пример предложения для получения эмбеддинга."

# Получение эмбеддинга предложения
embedding = get_sentence_embedding(sentence, tokenizer, model)
print(embedding.size())
````
